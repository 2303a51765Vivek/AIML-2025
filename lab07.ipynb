```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # Import the SVC class from the sklearn.svm module

# Define a dataframe with only features
df_feat = df.drop('Cancer',axis=1)
#df_feat.head()
df_feat.shape

# Define a dataframe with only target results for
# cancer detection
df_target = df['Cancer']
#df_target.head()
df_target.shape

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_feat, df_target, test_size=0.3, random_state=101) # Adjust test_size and random_state as needed

model = SVC()

# Now you can fit the model
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix # Import the confusion_matrix function

predictions = model.predict(X_test)
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score # Import accuracy_score

predictions = model.predict(X_test)
accuracy_score(y_test, predictions) # Calculate and print accuracy
```

```python
from sklearn.metrics import classification_report # Import classification_report

predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
```

```python
print("Test Data Misclassification error rate:",
      round(np.mean(predictions!=y_test),3))

train_pred = model.predict(X_train)
print("Train Data Misclassification error rate:",
      round(np.mean(train_pred!=y_train),3))
print(f"Accuracy of train data:{accuracy_score(y_train, train_pred)}")
print(f"Accuracy of test data:{accuracy_score(y_test, predictions)}")
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000],
              'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# Add the underscore
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
